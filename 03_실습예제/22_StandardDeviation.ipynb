```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

prior_mu = 0.0
prior_sigma = 0.5
x = np.linspace(-8, 12, 1000)
prior_y = (1/np.sqrt(2* np.pi*prior_sigma**2))* np.exp(-(x-prior_mu)**2 / (2* prior_sigma**2))

measured_mu = 4.0
measured_sigma = 0.9
measured_y = (1/np.sqrt(2* np.pi*measured_sigma**2))* np.exp(-(x-measured_mu)**2 / (2* measured_sigma**2))

product_mu = (prior_mu*(measured_sigma**2)+measured_mu*(prior_sigma**2))/(prior_sigma**2+measured_sigma**2)
product_sigma =(prior_sigma**2 * measured_sigma**2) / (prior_sigma**2+measured_sigma**2)
product_y = (1/np.sqrt(2* np.pi*product_sigma**2))* np.exp(-(x-product_mu)**2 / (2* product_sigma**2))

plt.style.use('default')
plt.rcParams['figure.figsize'] = (6,3)
plt.rcParams['font.size'] = 12
plt.rcParams['lines.linewidth'] = 3

plt.plot(x, prior_y, alpha=0.7, label='prior of N(0, 0,5)')
plt.plot(x, measured_y, alpha=0.7, label='Measure of N(4, 0,9)')
plt.plot(x, product_y, alpha=0.7, label='product')
plt.show()
```
